```python
import requests
from bs4 import BeautifulSoup
url="https://www.allsides.com/media-bias/media-bias-ratings"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
data=soup.find('table',{"class":"views-table cols-4"})
data
```

```python
heads=data.find_all('th')
heading=[h.get_text().strip() for h in heads][:11]
heading
```

```python
table_data=data.find_all('tr')
table_data
```

```python
table_d=data.find_all(class_="odd views-row-first")
odd=[b.get_text().strip().split()for b in table_d]
odd
```

```python
table_d=data.find_all(class_="even")
even=[b.get_text().strip().split()for b in table_d]
even
```
