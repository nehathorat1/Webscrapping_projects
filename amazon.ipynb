```python
import requests
from bs4 import BeautifulSoup
url="https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_2?ie=UTF8&pg=1"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
#soup
```

```python
all_books=soup.find_all(class_="zg-item-immersion")
#all_books
```

```python
first=all_books[0]
first
```

```python
f_title=first.find(class_="a-link-normal").get_text().strip()
f_title
```

```python
f_author=first.find(class_="a-size-small a-link-child").get_text().strip()
f_author
```

```python
f_price=first.find(class_="a-link-normal a-text-normal").get_text().strip().split('\n')[0].lstrip('₹\xa0')
f_price
```

```python
import pandas as pd
d={'titles':[],'authors':[],'prices':[]}
for i in range(2):
    url="https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_2?ie=UTF8&pg=1"
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html5lib')
    all_books=soup.find_all(class_="zg-item-immersion")
    for f in all_books:
        d['titles'].append(f.find(class_="a-link-normal").get_text().strip())
        d['authors'].append(f.find(class_="a-size-small a-link-child"))
        d['prices'].append(f.find(class_="a-link-normal a-text-normal").get_text().strip().split('\n')[0].lstrip('₹\xa0'))
df=pd.DataFrame(d)
df.head(50)
```
