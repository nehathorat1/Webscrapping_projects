```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://www.delhisldc.org/Loaddata.aspx?mode=14/01/2021"
page=requests.get(url)
soup=BeautifulSoup(page.content,'lxml')
soup
```

```python
data=soup.find(id="ContentPlaceHolder3_DGGridAv")
data
```

```python
rows=data.find_all('tr')
rows[0]
```

```python
heads=[p.get_text() for p in rows[0].find_all('td')]
heads
```

```python
table=[]
for d in rows[1:]:
    table.append([p.get_text() for p in d.find_all('td')])
table
```

```python
dict={}
for k in heads:
    dict[k]=[]
dict
```

```python
for t in table:
    dict['TIMESLOT'].append(t[0])
    dict['DELHI'].append(t[1])
    dict['BRPL'].append(t[2])
    dict['BYPL'].append(t[3])
    dict['NDPL'].append(t[4])
    dict['NDMC'].append(t[5])
    dict['MES'].append(t[6])
dict
```

```python
df=pd.DataFrame(dict)
df.to_csv('data/14_01_2021.csv',index=False)
```
