```python
import requests
from bs4 import BeautifulSoup
url="https://stackoverflow.com/questions?tab=newest&page=1"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html')
soup
```

```python
all_que=soup.find(id="questions")
all_que
```

```python
ques=all_que.find_all(class_="question-summary")
ques
```

```python
first=ques[0]
first
```

```python
fq=first.find(class_="question-hyperlink").get_text().strip()
fq
```

```python
fs=first.find(class_="excerpt").get_text().strip('\r\n ')
fs
```

```python
import pandas as pd
d={'question':[],'description':[]}
for i in range(1,50,15):
    url=f"https://stackoverflow.com/questions?tab=newest&page={i}"
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html')
    all_que=soup.find_all(class_="question-summary")
    for f in all_que:
        d['question'].append(f.find(class_="question-hyperlink").get_text().strip())
        d['description'].append(f.find(class_="excerpt").get_text().strip('\r\n '))
df=pd.DataFrame(d)
df.head(55)
```
