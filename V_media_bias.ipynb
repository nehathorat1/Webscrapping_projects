```python
import requests
from bs4 import BeautifulSoup
url="https://www.allsides.com/media-bias/media-bias-ratings"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html')
#soup
```

```python
data=soup.find(class_="views-table cols-4")
#data
```

```python
heads=soup.find_all("th")
heading=[h.get_text().strip() for h in heads]
heading
```

```python
table_body=data.find_all('tr')
len(table_body)
```

```python
td1=data.find_all(class_='odd')
odnew=[b.get_text().strip().split() for b in td1]
odnew
```

```python
td3=data.find_all(class_='even')
evnew=[a.get_text().strip().split() for a in td3]
#evnew
```

```python
d={}
for h in heading:
    d[h]=[]
d
```

```python
for k in odnew: 
    #if len(k)==6:
        d['News Source'].append(k[:1])
        d['AllSides Bias Rating'].append(k[5:9])
        d['What do you think?'].append(k[2:5])
        d['Community feedback (biased, not normalized)'].append(k[9:])
print(d)
```

```python
d1={}
for h in heading:
    d1[h]=[]
d1
for j in evnew: 
    #if len(j)==6:
        d1['News Source'].append(j[0])
        d1['AllSides Bias Rating'].append(j[1])
        d1['What do you think?'].append(j[2])
        d1['Community feedback (biased, not normalized)'].append(j[3])
print(d1)
```
